```python
# REVISION
# =========================================
# 1. IMPORT
# =========================================
import numpy as np

# =========================================
# 2. CREATE ARRAYS
# =========================================

arr1 = np.array([1, 2, 3])              # 1D array
arr2 = np.array([[1, 2], [3, 4]])      # 2D array

np.zeros((2,3))   # all zeros
np.ones((2,3))    # all ones
np.eye(3)         # identity matrix

# =========================================
# 3. ARRAY PROPERTIES
# =========================================

print(arr2.shape)   # (rows, cols)
print(arr2.ndim)    # dimensions
print(arr2.size)    # total elements

# =========================================
# 4. INDEXING & SLICING
# =========================================

arr = np.array([10, 20, 30, 40])

print(arr[0])       # first element
print(arr[1:3])     # slicing

# 2D indexing
arr2 = np.array([[1,2],[3,4]])
print(arr2[0,1])    # row 0, col 1

# =========================================
# 5. ARRAY OPERATIONS (VECTORIZED)
# =========================================

a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)   # element-wise addition
print(a * b)   # element-wise multiplication

# =========================================
# 6. BROADCASTING
# =========================================

print(a + 5)   # add scalar to all elements

# =========================================
# 7. BASIC FUNCTIONS
# =========================================

arr = np.array([1,2,3,4])

print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))

# =========================================
# 8. RESHAPING
# =========================================

arr = np.array([1,2,3,4,5,6])
print(arr.reshape(2,3))   # reshape

# =========================================
# 9. RANDOM
# =========================================

np.random.rand(3)           # random floats
np.random.randint(1,10,5)   # random integers

# =========================================
# 🔴 ADVANCED NUMPY
# =========================================

# =========================================
# 10. BOOLEAN INDEXING
# =========================================

arr = np.array([10,20,30,40])
print(arr[arr > 20])   # filter values

# =========================================
# 11. FANCY INDEXING
# =========================================

print(arr[[0,2]])   # select multiple indices

# =========================================
# 12. AXIS OPERATIONS
# =========================================

arr = np.array([[1,2],[3,4]])

print(np.sum(arr))         # total
print(np.sum(arr, axis=0)) # column-wise
print(np.sum(arr, axis=1)) # row-wise

# =========================================
# 13. FLATTEN
# =========================================

arr = np.array([[1,2],[3,4]])
print(arr.flatten())   # convert to 1D

# =========================================
# 14. STACKING
# =========================================

a = np.array([1,2])
b = np.array([3,4])

print(np.vstack((a,b)))   # vertical
print(np.hstack((a,b)))   # horizontal

# =========================================
# 15. LINEAR ALGEBRA
# =========================================

a = np.array([1,2])
b = np.array([3,4])

print(np.dot(a,b))   # dot product

A = np.array([[1,2],[3,4]])
B = np.array([[5,6],[7,8]])

print(np.matmul(A,B))   # matrix multiplication
print(A.T)              # transpose

# =========================================
# 16. BROADCASTING (ADVANCED)
# =========================================

arr = np.array([[1,2,3],[4,5,6]])
print(arr + np.array([1,1,1]))   # row-wise addition

# =========================================
# 17. WHERE
# =========================================

arr = np.array([10,20,30,40])
print(np.where(arr > 20, "High", "Low"))

# =========================================
# 18. UNIQUE & SORT
# =========================================

arr = np.array([3,1,2,3,2])
print(np.unique(arr))
print(np.sort(arr))

# =========================================
# 19. PERFORMANCE TEST
# =========================================

import time

arr = np.arange(1000000)

start = time.time()
arr * 2
print("NumPy time:", time.time() - start)

# =========================================
# 20. MATRIX MULTIPLICATION RULE FIX (IMPORTANT)
# =========================================

arr = np.array([[2,5,6],[5,7,2]])  # shape (2x3)

print(np.sum(arr, axis=0))  # column sum

# FIX (use transpose)
print(np.matmul(arr, arr.T))  # valid (2x3)*(3x2)

# =========================================
# DONE ✅
# =========================================
```
