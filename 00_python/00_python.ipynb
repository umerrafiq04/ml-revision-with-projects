```python
# ================================
# 1. BASICS
# ================================

name = "Umer"
age = 21
print(f"My name is {name}, age {age}")

# Input
# user = input("Enter name: ")

# Operators
a, b = 10, 3
print(a+b, a-b, a*b, a/b, a//b, a%b)

# Condition
if age > 18:
    print("Adult")
else:
    print("Minor")

# Loop
for i in range(3):
    print(i)

# ================================
# 2. DATA STRUCTURES
# ================================

# String
text = "python"
print(text[::-1])  # reverse

# List
nums = [1,2,3,4]
nums.append(5)
even = [x for x in nums if x % 2 == 0]

# Tuple
t = (1,2,3)

# Set
s = {1,2,2,3}
print(s)

# Dictionary
student = {"name":"Umer", "marks":90}
for k,v in student.items():
    print(k,v)

# ================================
# 3. FUNCTIONS
# ================================

def add(a,b):
    return a+b

print(add(2,3))

# lambda
square = lambda x: x*x
print(square(5))

# map, filter
nums = [1,2,3,4]
print(list(map(lambda x:x*2, nums)))
print(list(filter(lambda x:x%2==0, nums)))

# reduce
from functools import reduce
print(reduce(lambda a,b:a+b, nums))

# recursion
def fact(n):
    return 1 if n==1 else n*fact(n-1)

print(fact(5))

# ================================
# 4. FILE HANDLING
# ================================

with open("test.txt","w") as f:
    f.write("Hello")

with open("test.txt","r") as f:
    print(f.read())

# JSON
import json
data = {"name":"Umer"}
with open("data.json","w") as f:
    json.dump(data,f)

# ================================
# 5. OOP
# ================================

class Student:
    college = "KU"

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(self.name, self.marks)

s1 = Student("Umer", 90)
s1.display()

# Inheritance
class Person:
    def greet(self):
        print("Hello")

class Child(Person):
    pass

Child().greet()

# Polymorphism
class Dog:
    def sound(self):
        print("Bark")

class Cat:
    def sound(self):
        print("Meow")

for a in [Dog(), Cat()]:
    a.sound()

# ================================
# 6. ADVANCED PYTHON
# ================================

# Iterator
nums = [1,2,3]
it = iter(nums)
print(next(it))
print(next(it))

# Generator
def gen(n):
    for i in range(n):
        yield i

for x in gen(3):
    print(x)

# Decorator
def deco(func):
    def wrapper():
        print("Before")
        func()
        print("After")
    return wrapper

@deco
def hello():
    print("Hello")

hello()

# Exception
try:
    x = int("abc")
except:
    print("Error")

# ================================
# 7. REAL WORLD
# ================================

# API
import requests
res = requests.get("https://api.github.com")
print(res.status_code)

# ================================
# 8. ITER INTERNAL (IMPORTANT)
# ================================

nums = [1,2,3]
it = iter(nums)

while True:
    try:
        x = next(it)
        print(x)
    except StopIteration:
        break

# ================================
# DONE ✅
# ================================
```
