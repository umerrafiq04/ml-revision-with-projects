```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.DataFrame({
    "day": [1, 2, 3, 4, 5],
    "sales": [100, 150, 130, 170, 200]
})

sns.lineplot(x="day", y="sales", data=df)
plt.title("Sales over Days")
plt.show()
```

```python
df = pd.DataFrame({
    "student": ["A", "B", "C", "D"],
    "marks": [80, 90, 75, 88]
})

sns.barplot(x="student", y="marks", data=df)
plt.title("Student Marks")
plt.show()
```

```python
df = pd.DataFrame({
    "marks": [50, 60, 65, 70, 75, 80, 85, 90, 95, 100]
})

sns.histplot(df["marks"], bins=5)
plt.title("Marks Distribution")
plt.show()
```

```python
df = pd.DataFrame({
    "hours": [1, 2, 3, 4, 5, 6],
    "marks": [45, 50, 60, 65, 70, 80]
})

sns.scatterplot(x="hours", y="marks", data=df)
plt.title("Hours vs Marks")
plt.show()
```

```python
df = pd.DataFrame({
    "group": ["A", "A", "A", "B", "B", "B"],
    "marks": [70, 75, 90, 60, 65, 95]
})

sns.boxplot(x="group", y="marks", data=df)
plt.title("Box Plot")
plt.show()
```

```python
df = pd.DataFrame({
    "department": ["CS", "IT", "CS", "ECE", "IT", "CS"]
})

sns.countplot(x="department", data=df)
plt.title("Department Count")
plt.show()
```

```python
df = pd.DataFrame({
    "math": [80, 90, 70, 60],
    "science": [85, 88, 75, 65],
    "english": [78, 92, 72, 68]
})

corr = df.corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
iris = sns.load_dataset("iris")
sns.pairplot(iris)
plt.show()
```

```python
iris = sns.load_dataset("iris")

sns.scatterplot(
    x="sepal_length",
    y="sepal_width",
    hue="species",
    data=iris
)
plt.title("Sepal Length vs Width")
plt.show()
```
