```python
# REVISION
# =========================================
# 1. IMPORT
# =========================================
import pandas as pd

# =========================================
# 2. CREATE DATAFRAME
# =========================================

data = {
    "name": ["Umer", "Ali", "Sara", "John"],
    "age": [21, 22, 20, 23],
    "marks": [85, 90, 78, 88]
}

df = pd.DataFrame(data)
print(df)

# =========================================
# 3. BASIC INFO
# =========================================

print(df.head())     # first rows
print(df.tail())     # last rows
print(df.shape)      # (rows, cols)
print(df.columns)    # column names
print(df.info())     # full info
print(df.describe()) # stats

# =========================================
# 4. SELECT DATA
# =========================================

print(df["name"])                 # single column
print(df[["name", "marks"]])      # multiple columns

print(df.iloc[0])                 # row by index
print(df.iloc[0,1])               # specific value

# =========================================
# 5. FILTERING (VERY IMPORTANT)
# =========================================

print(df[df["marks"] > 80])

print(df[(df["marks"] > 80) & (df["age"] > 21)])

# =========================================
# 6. ADD / MODIFY COLUMN
# =========================================

df["grade"] = ["B", "A", "C", "B"]

df["marks"] = df["marks"] + 5

# =========================================
# 7. DROP
# =========================================

df.drop("age", axis=1, inplace=True)

# =========================================
# 8. SORTING
# =========================================

print(df.sort_values("marks", ascending=False))

# =========================================
# 9. GROUPBY (VERY IMPORTANT)
# =========================================

print(df.groupby("grade")["marks"].mean())

# multiple aggregation
print(df.groupby("grade")["marks"].agg(["mean","max","min"]))

# =========================================
# 10. HANDLE MISSING VALUES
# =========================================

df.isnull()
df.fillna(0)
df.dropna()

# =========================================
# 11. APPLY
# =========================================

df["marks"] = df["marks"].apply(lambda x: x + 2)

# =========================================
# 12. VALUE COUNTS
# =========================================

print(df["grade"].value_counts())

# =========================================
# 13. MIN / MAX ROW (IMPORTANT)
# =========================================

print(df.loc[df["marks"].idxmin()])     # best method
print(df.nsmallest(1, "marks"))

# =========================================
# 14. MERGE / JOIN (VERY IMPORTANT)
# =========================================

df1 = pd.DataFrame({
    "id": [1,2],
    "name": ["Umer","Ali"]
})

df2 = pd.DataFrame({
    "id": [1,2],
    "marks": [90,85]
})

merged = pd.merge(df1, df2, on="id")
print(merged)

# =========================================
# 15. PIVOT TABLE
# =========================================

print(pd.pivot_table(df, values="marks", index="grade", aggfunc="mean"))

# =========================================
# 16. CORRELATION (ML IMPORTANT)
# =========================================

# print(df.corr())

# =========================================
# 17. SAVE & LOAD
# =========================================

df.to_csv("data.csv", index=False)

df = pd.read_csv("data.csv")

# =========================================
# 🔥 REAL WORKFLOW (ML PIPELINE STYLE)
# =========================================

df = pd.read_csv("data.csv")

df.dropna(inplace=True)
df.drop_duplicates(inplace=True)

df["grade"] = df["marks"].apply(lambda x: "A" if x > 80 else "B")

print(df.groupby("grade")["marks"].mean())

print(df.sort_values("marks", ascending=False))

# =========================================
# DONE ✅
# =========================================
```
