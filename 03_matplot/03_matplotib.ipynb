```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[2,4,6,8,10]
plt.plot(x,y)
plt.show()
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x, y)

plt.title("Simple Line Graph")
plt.xlabel("X values")
plt.ylabel("Y values")

plt.show()
```

```python
plt.plot(x, y, color='red', marker='o', linestyle='--')

plt.title("Styled Graph")
plt.show()
```

```python
y2 = [5, 15, 20, 2,5]

plt.plot(x, y, label="Line 1")
plt.plot(x, y2, label="Line 2")

plt.legend()
plt.show()
```

```python
names = ["A", "B", "C"]
values = [5, 7, 3]

plt.bar(names, values)
plt.title("Bar Chart")
plt.show()
```

```python
data = [1,2,2,3,3,3,4,5,5,6]

plt.hist(data, bins=9)
plt.title("Histogram")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [10,20,25,30,50]

plt.scatter(x, y,color='red',marker="o")
plt.title("Scatter Plot")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y)
plt.title("Line")

plt.subplot(1,2,2)
plt.bar(names, values)
plt.title("Bar")

plt.show()
```
